```python
df.columns = (
    df.columns
    .astype(str)
    .str.strip()
    .str.lower()
    .str.replace(" ", "")
)

print(df.columns)
```

Problem Statement:
The objective of this project is to analyze online gaming behavior
using player demographic details, game preferences, playtime,
sessions, achievements, and engagement levels with the help of Python.


Dataset Description:
The dataset contains information about online game players including
their age, gender, location, game genre, playtime, sessions per week,
achievements, and engagement level. The dataset is used to identify
patterns in gaming behavior.


Data Cleaning

```python
# Missing values
df.isnull().sum()
```

```python
# Duplicate rows
df.duplicated().sum()
df = df.drop_duplicates()

```

```python
def find_col(keys):
    for c in df.columns:
        for k in keys:
            if k in c:
                return c
    return None

col_gender   = find_col(["gender"])
col_genre    = find_col(["genre"])
col_playtime = find_col(["playtime"])
col_achieve  = find_col(["achieve"])
col_engage   = find_col(["engage"])
col_session  = find_col(["session"])

print(col_gender, col_genre, col_playtime, col_achieve, col_engage, col_session)
```

```python
df["playtime_hours"] = pd.to_numeric(df[col_playtime], errors="coerce")
df["achievements"]  = pd.to_numeric(df[col_achieve], errors="coerce")
df["sessions"]      = pd.to_numeric(df[col_session], errors="coerce")

df["gender"]     = df[col_gender]
df["genre"]      = df[col_genre]
df["engagement"] = df[col_engage]

df = df.dropna(subset=["playtime_hours", "achievements", "engagement"])
df.head()
```

EXPLORATORY DATA ANALYSIS (AWP CORE)

Gender Distribution


```python
sns.countplot(data=df, x="gender")
plt.title("Gender Distribution of Players")
plt.show()
```

Game Genre Distribution

```python
sns.countplot(data=df, x="genre")
plt.title("Game Genre Distribution")
plt.show()
```

Engagement Level Distribution

```python
sns.countplot(data=df, x="engagement")
plt.title("Engagement Level Distribution")
plt.show()
```

Average Playtime by Game Genre

```python
sns.barplot(data=df, x="genre", y="playtime_hours")
plt.title("Average Playtime by Game Genre")
plt.show()
```

Playtime vs Achievements

```python
sns.scatterplot(
    data=df,
    x="playtime_hours",
    y="achievements",
    hue="engagement"
)
plt.title("Playtime vs Achievements")
plt.show()
```

Sessions vs Engagement Level

```python
sns.boxplot(data=df, x="engagement", y="sessions")
plt.title("Sessions per Week vs Engagement Level")
plt.show()
```

Key Insights

1. Male players dominate the gaming population.
2. Action and Strategy games are the most preferred genres.
3. Players with higher engagement levels play more frequently.
4. Increased playtime leads to higher achievements.
5. Engagement level strongly affects player activity.
